## Introduction
Financial losses happening every year due to credit card fraud transactions have forced the financial industry to swtich from a posterior investigation approach to an priori predictive approach which uses fraud detection algorithms to identify possible fraudsters and take proper measures before lending any amount to them.

The credit giving organizations find it hard to give credit to individuals because of their lacking or missing record as a consumer. A few customers utilize this for their potential benefit by turning into a defaulter. If someone works for a purchaser finance organization that has some expertise in loaning different kinds of credits to clients, he should utilize Exploratory Data Analysis (EDA) to examine the examples present in the information which will ensure that the loans are not dismissed for the candidates fit for reimbursing.

At the point when the organization gets a credit application, the organization needs to privileges for advance endorsement in view of the candidate's profile. Two kinds of dangers are related with the bank's or alternately organization's choice:

i) In the event that the competitor is probably going to reimburse the credit, not endorsing the credit tends in a business misfortune to the organization.

ii) In the event that the applicant isn't probably going to reimburse the advance, for example he/she is probably going to default/extortion, then, at that point, supporting the credit might prompt a monetary misfortune for the organization.

At the point when a client applies for a credit, there are four kinds of choices that could be taken by the bank/organization:

i) Approved

ii) Cancelled

iii) Refused

iv) Unused offer: The applicant cancelled loan application but at different stages of the process.

In this study, we will try to understand how consumer attributes and loan attributes impact the tendency of default using Exploratory Data Analysis(EDA).

### Creating function to plot graphs for all the Catogrical columns WRT Target 0 and Target 1

```python
def categorical_plot(var):
    plt.figure(figsize=(40,20))
    
    plt.rcParams['axes.labelpad'] = 50
    plt.subplot(1, 2, 1)
    sns.countplot(var, data=Target0, palette = 'Set3', hue='TARGET') 
    plt.xlabel(var, fontsize= 30, fontweight="bold")                                                         #Target 0
    plt.ylabel('Non Payment Difficulties', fontsize= 30, fontweight="bold")
    plt.xticks(rotation=90, fontsize=30)
    plt.yticks(rotation=360, fontsize=30)
    
    
    plt.rcParams['axes.labelpad'] = 50
    plt.subplot(1, 2, 2)
    sns.countplot(var, data=Target1, palette = 'Set1', hue='TARGET')                                           # Target 1
    plt.xlabel(var, fontsize= 30, fontweight="bold")
    plt.ylabel('Payments Difficulties', fontsize= 30, fontweight="bold")
    plt.xticks(rotation=90, fontsize=30)
    plt.yticks(rotation=360, fontsize=30)
    plt.show()
```
